```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2) #  2 
pca.fit(digits.data)
```

```python
digits_pca = pca.transform(digits.data)
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525",
          "#A83683", "#4E655E", "#853541", "#3A3120","#535D8E"]
```

```python
plt.figure(figsize=(10, 10))

#     x, y .
plt.xlim(digits_pca[:, 0].min(), digits_pca[:, 0].max())
plt.ylim(digits_pca[:, 1].min(), digits_pca[:, 1].max())


for i in range(len(digits.data)):
    pca1 = digits_pca[i,0] #  
    pca2 = digits_pca[i,1] #  
    col_num = colors[digits.target[i]]
    #     
    plt.text(pca1, pca2, str(digits.target[i]),
             color = col_num,
             fontdict={'weight': 'bold', 'size': 9})
    
plt.xlabel("  ")
plt.ylabel("  ")
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=42)
d_tsne = tsne.fit_transform(digits.data)

print("   : ", digits.data.shape)
print("tsne    : ", d_tsne.shape)
```

```python
plt.figure(figsize=(10,10)) #   .
plt.xlim(d_tsne[:,0].min(), d_tsne[:,0].max() + 1)
plt.ylim(d_tsne[:,1].min(), d_tsne[:,1].max() + 1)
for i in range(len(digits.data)):
    #     .
    plt.text(d_tsne[i,0], 
             d_tsne[i,1],
             str(digits.target[i]),
             color =  colors[digits.target[i]],
             fontdict={'weight':'bold', 'size':9})
plt.xlabel("t-SNE  0")
plt.xlabel("t-SNE  1")
```

```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
people = fetch_lfw_people(min_faces_per_person=20, 
                          resize=0.7) #   color=True  .

print(people.keys())
print("  ", people.images.shape)
print(" " , people.target_names) #  62
image_shape = people.images[0].shape # 3023 

print(image_shape)
```

```python
fig, axes = plt.subplots(2, 5, figsize=(15,8),
                        subplot_kw= {'xticks':(), 'yticks':() })

for target, image, ax in zip(people.target, people.images, axes.ravel() ):
    ax.imshow(image) #   
    ax.set_title(people.target_names[target]) #  
```

```python
print(" :", people.images.shape)
print(" :", len(people.target_names))
```

```python
#     
counts = np.bincount(people.target)
print( counts )
print(" : {},  : {}".format(np.min(counts), np.max(counts)))
```

```python
nums = people.target.shape  #  
mask = np.zeros(nums, dtype=np.bool)
mask
```

```python
np.unique(people.target)
```

```python
#   50 
nums = np.unique(people.target)

for target in nums:
    mask[np.where(people.target==target)[0][:50]] = 1 #   50 

#   1    
X_people = people.data[mask]
y_people = people.target[mask]

# 0 ~ 255      0~1  .
# () MinMaxScaler    .
X_people = X_people / 255.
```

```python
from sklearn.neighbors import KNeighborsClassifier
X_train, X_test, y_train, y_test = train_test_split(
        X_people, y_people, stratify=y_people, random_state=0)

#      KNeightborsClassifier  .
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("1-    :{:.2f}".format(knn.score(X_test, y_test)))
```

```python
from sklearn.decomposition import PCA
```

```python
#   PCA      . 
# PCA    0     PCA  StandardScaler   .
pca = PCA(n_components=100, whiten=True, random_state=0).fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

print("X_train_pca.shape:", X_train_pca.shape)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_pca, y_train)

print("  : {:.2f}".format(knn.score(X_test_pca, y_test)))
```

```python
print("pca.components_.shape ", pca.components_.shape)

fig, axes = plt.subplots(3, 5, figsize=(15,12),
                        subplot_kw={'xticks':(), 'yticks':() })

for i, (component, ax) in enumerate(zip(pca.components_, axes.ravel())):
    ax.imshow(component.reshape(image_shape), cmap="viridis")
    ax.set_title(" {}".format((i+1)))
```
