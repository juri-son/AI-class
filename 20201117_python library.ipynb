## numpy : 딥러닝에서 많이 쓰이는 라이브러리.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,14,200)
x
```

```python
# x의 값을 이용하여 y값 (sin) 구하기
y = np.sin(x)

# 그래프 그리기
plt.plot(x,y,'o')
```

```python
#cos값 그리기
y = np.cos(x)
plt.plot(x,y,'^')
```

```python
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.tan(x)

plt.plot(x,y1, label = "cos(x)")
plt.plot(x,y2, label = "sin(x)")
plt.plot(x,y3, label = "tan(x)")
plt.legend()

plt.title("sin, cos, tan plot")
```
